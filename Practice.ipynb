```python
import pandas as pd
```

```python
df=pd.read_csv("practice.csv")
```

```python
df
```

```python
import matplotlib.pyplot as pt
from matplotlib  import style 
x1=df['Age'].values
x2=df['Job'].values



pt.plot(x1,x2,color="r")
pt.title("data")
pt.xlabel("Age")
pt.ylabel("Job")

pt.show()
```

```python
import matplotlib.pyplot as pt
from matplotlib  import style 
x1=df['Age'].values

pt.pie(x1)




pt.show()
```

```python
import matplotlib.pyplot as pt
from matplotlib  import style 
x1=df['Age'].values
x2=df['Job'].values



pt.bar(x1,x2,color="r")
pt.title("data")
pt.xlabel("Age")
pt.ylabel("Job")

pt.show()
```
