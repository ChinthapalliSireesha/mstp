```python
x1=[1,2,3,4,5]
x2=[1,4,6,8,9]

y=[1,2,3,4,5]
fig=pt.figure()
a=fig.add_axes([0,0,1,1])
pt.title("Sample",color='g')
pt.xlabel("distance")
pt.ylabel("time")
pt.plot(x1,y,color="g")
pt.plot(x2,y,color="y")
a.legend(labels=(x1,x2),loc="lower right")
```

```python
x=[1,2,3,4,5]
y=['A','B','C','D','E']
fig=pt.figure()
a=fig.add_axes([0,0,1,1])
pt.xlabel("distance")
pt.ylabel("time")
pt.pie(x,labels=y,autopct='%1.2f%%')
pt.show()
```
